```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5, 11)
y = x ** 2
```

```python
plt.plot(x, y)
plt.xlabel('X LABEL')
plt.ylabel('Y LABEL')
plt.title('Title')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('X LABEL')
axes.set_ylabel('Y LABEL')
axes.set_title('Title')
```
