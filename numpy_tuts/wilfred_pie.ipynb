```python
import numpy as np 

postive_numbers = np.arange(1, 1000, 4)
postive_numbers
```

```python
positive_numbers_as_fraction = 1 / (postive_numbers)
positive_numbers_as_fraction
```

```python
negative_numbers_as_fraction = 1 / (postive_numbers + 2)
negative_numbers_as_fraction
```

```python
pie_number = 4 * (sum(positive_numbers_as_fraction) - sum(negative_numbers_as_fraction))
pie_number
```
